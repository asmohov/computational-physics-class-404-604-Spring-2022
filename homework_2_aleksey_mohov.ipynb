```python
import numpy as np
import matplotlib,pylab as p

#set up linspace for plotting; these are the inputs to our function
x_vals = np.arange(-1*np.pi,np.pi,.01)

def func(i,en):
    out=np.pi/2
    for k in range(0,en+1): #iterate over wavenumbers from 0 to n, en to distinguish from later used n
        if k%2 == 1: #select only odd k's
            out = out+((2*np.sin(k*i))/k) #evaluate the requested function for a given n, i
    return out
n_list = (11,21,31)


for n in n_list:

    y_vals = [] #reset y_values for each plot

    for x in x_vals: #iterate over x domain
        y_vals = np.append(y_vals,func(x,n))
    p.plot(x_vals,y_vals)
p.show()
```
